```python
plt.figure(figsize=(12,6))
top_teams = df.groupby("TEAM")["W"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_teams.index, y=top_teams.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Top 10 Teams by Wins")
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8,6))

# Boxplot with attractive colors
sns.boxplot(
    x=df["ADJOE"],
    color="#FF6F61",     # Coral, bright and attractive
    fliersize=6,          # slightly bigger outliers
    linewidth=2,          # thicker box edges
    medianprops=dict(color="#2E8B57", linewidth=2)  # SeaGreen median line
)

# Title and labels
plt.title("Distribution of Adjusted Offensive Efficiency (ADJOE)", fontsize=16, fontweight="bold", color="#333333")
plt.xlabel("ADJOE", fontsize=12)

plt.tight_layout()
plt.show()
```

### Insights
- **Box Plot shows the distribution of Offensive Efficiency (ADJOE)** across all teams.  
- The **median ADJOE is around 105**, meaning most teams score near this efficiency level.  
- The **interquartile range (IQR)** lies between ~100 and ~110, showing the central spread.  
- **Outliers exist** on both ends:
  - Teams with **very low ADJOE (~80‚Äì90)** represent weak offensive performers.  
  - Teams with **very high ADJOE (~120‚Äì130)** represent exceptionally strong offensive teams.  
- This suggests that while most teams cluster near average efficiency, a few outliers **stand out significantly in performance**.

### Insights ‚Äì KDE Plot (Wins Distribution)
- Most teams cluster around **10‚Äì20 wins per season**.  
- **Few teams exceed 25+ wins**, showing how difficult it is to maintain elite performance.  
- The KDE curve confirms that the **majority of teams are average performers**, with only a small tail of top teams.  


```python
plt.figure(figsize=(12,6))
sns.violinplot(x="CONF", y="W", data=df, palette="Set2")
plt.xticks(rotation=90)
plt.title("Wins by Conference")
plt.show()
# Small title
print("\nTop 10 Winning Teams in ACC, B12, and B10 Conferences\n")

# Top teams by Wins in ACC, B12, B10
top_teams = df[df['CONF'].isin(['ACC','B12','B10'])].sort_values(by='W', ascending=False)
print(top_teams[['TEAM','CONF','W']].head(10))
```

### Insights ‚Äì Violin Plot (Wins by Conference)
- Conferences like **ACC, B12, and B10** show a **higher concentration of winning teams**.  
- Smaller conferences have **greater variability** and fewer consistently high-win teams.  
- The spread highlights **competitive imbalance** between large vs. small conferences.  


```python
plt.figure(figsize=(14,6))
sns.set_style("whitegrid")  # clean background

# Strip plot with color for each conference
sns.stripplot(
    x="CONF",
    y="ADJOE",
    data=df,
    size=5,               # slightly bigger points for visibility
    jitter=True,
    alpha=0.7,
    palette="Set2"        # attractive colors
)

# Optional: add boxplot behind stripplot for summary stats
sns.boxplot(
    x="CONF",
    y="ADJOE",
    data=df,
    whis=[0,100],         # show full range
    width=0.3,
    showcaps=False,
    boxprops={'facecolor':'None'},  # transparent boxes
    showfliers=False,
    whiskerprops={'linewidth':0}
)

plt.xticks(rotation=45, fontsize=10)
plt.ylabel("Adjusted Offensive Efficiency (ADJOE)", fontsize=12)
plt.xlabel("Conference", fontsize=12)
plt.title("Offensive Efficiency (ADJOE) by Conference", fontsize=16)
plt.tight_layout()
plt.show()
```

### Insights ‚Äì Strip Plot (Offensive Efficiency by Conference)
- **ACC, B12, and B10** conferences cluster toward **higher ADJOE values**, confirming stronger overall offenses.  
- Smaller conferences show **wider spread and inconsistency**.  
- Outliers highlight individual teams that **outperform or underperform** compared to their peers.
